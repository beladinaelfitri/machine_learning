```python
import pandas as pd
import numpy as np
import copy
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/machine-learning-course/syllabus/gh-pages/hiw-2019b/dataset-students-ml-2019b.csv')
df['nim'] = df['nim'].astype(str)

#encoding dof
dof_enc = {"dow_of_birth" : {"Monday":1, "Tuesday":2, "Wednesday":3, "Thursday":4, "Friday":5, "Saturday":6, "Sunday":7}}
df.replace(dof_enc, inplace=True)
df
```

```python
#select features : dow_of_birth, height and weight
def getData(row):
    return [row['nim'],row['dow_of_birth'],row['height'],row['weight'],0,0]
```

```python
X = df.apply(getData, axis=1).tolist()
X
```

```python
#select k = 4
k = 4

#select random point as centroids
centroids = []
for i in range(k):
    rand = np.random.randint(0, len(X)-1)
    centroids.append(X[rand][1:4])
```

```python
#calculate euclidian distance
def euclidian(u, v):
    return sum((p-q)**2 for p, q in zip(u, v))**0.5
```

```python
flag = 0
#perform k-means algorithm until all centroid point has same value
while (True):
    cluster = []
    flag += 1
    
    for i in range(len(X)):
        euclid = []
        #calculate euclidian distance
        euclid.append(euclidian(X[i][1:],centroids[0]))
        euclid.append(euclidian(X[i][1:],centroids[1]))
        euclid.append(euclidian(X[i][1:],centroids[2]))
        euclid.append(euclidian(X[i][1:],centroids[3]))
        #select cluster from minimum euclidean
        idx = np.argmin(euclid)
        cluster.append(idx+1)
        #add cluster and euclidian distance to X
        X[i][4] = idx+1
        X[i][5] = euclid[idx]
        
    #clustering the centroids
    group = {}
    for j in set(cluster):
        group[j] = [i for i in range(len(cluster)) if cluster[i] == j]

    #store centroid information of each cluster
    dataX = {}
    for j in range(1,5):
        dataX[j] = [X[group[j][i]][1:4] for i in range(len(group[j]))]

    #assign new centroids to each cluster
    new_centroids = []
    new_centroids.append(np.mean(dataX[1], axis=0).tolist())
    new_centroids.append(np.mean(dataX[2], axis=0).tolist())
    new_centroids.append(np.mean(dataX[3], axis=0).tolist())
    new_centroids.append(np.mean(dataX[4], axis=0).tolist())
    
    if (centroids == new_centroids):
        break
    
    centroids = copy.copy(new_centroids)
```

```python
print('Model Code: B')    
for i in range(len(X)):
    #print nim, cluster_number, euclidean distance to its centroid
    print('%s, %s, %0.4f' % (X[i][0], X[i][4], X[i][5]))
```
