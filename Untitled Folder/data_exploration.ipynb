# Data Exploration

```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 150
```

```python
df = pd.read_excel('students_record_2.xlsx')
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
mpl.rcParams['figure.dpi'] = 200
corr = df.select_dtypes(include = ['float64', 'int64']).iloc[:, :].corr()
plt.figure(figsize=(10, 10))
ax = sns.heatmap(corr, vmax=1, square=True)
# ax.set_xticklabels(rotation=30)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
mpl.rcParams['figure.dpi'] = 250
plt.figure(figsize=(60, 60))
f, axes = plt.subplots(1, 4)
sns.boxplot(y=df["height"], ax=axes[0])
sns.boxplot(y=df["weight"], ax=axes[1])
sns.boxplot(y=df["average study time"], ax=axes[2])
sns.boxplot(y=df["mid exam score"], ax=axes[3])
plt.subplots_adjust(wspace=1)
```

```python
sns.countplot(x = 'average study time', data = df)
```

```python
sns.distplot(df[df['mid exam score'].notnull()]['mid exam score'], kde=False, bins=10);
# sns.distplot(df['mid exam score'], kde=False, bins=10);
```

```python
df[df['mid exam score'].notnull()].describe()
```

```python
mpl.rcParams['figure.dpi'] = 250
plt.figure(figsize=(100, 100))
df_nona= df[df['mid exam score'].notnull()]
g = sns.pairplot(df_nona[['average study time', 'mid exam score']] )
```
